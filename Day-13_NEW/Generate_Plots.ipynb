```python
import pandas as pd

df = pd.read_csv("FAOSTAT_data_2-18-2022.csv")
df
```

```python
import matplotlib.pyplot as plt

df = df.sort_values(by='Value', ascending=False)
df = df[df["Value"] > 0.0]
fig = plt.figure(figsize=(18,7))
plt.scatter(df["Area"],df["Value"].to_list())
plt.xticks(rotation = 90,fontsize=10)
plt.yticks(fontsize=14)
plt.ylabel("Exported Cocoa Beans (Tonnes)", fontsize=16)
# plt.xlabel("Countries", fontsize=16)
plt.title("Cocoa Bean Exports by Country, 2018", fontsize=16)
plt.savefig("Linear_Plot_Cocoa_Beans_Export.jpg")
# plt.yscale("log")
# plt.show()
```
