```python
import pandas as pd

df = pd.read_csv("FAOSTAT_data_2-18-2022.csv")
df
```

```python
import matplotlib.pyplot as plt

df = df.sort_values(by='Value', ascending=False)
df = df[df["Value"] > 0.0]
fig = plt.figure(figsize=(18,10))
plt.scatter(df["Area"],df["Value"].to_list())
plt.xticks(rotation = 90) 
plt.ylabel("Exported Cocoa Beans (Tonnes)")
plt.xlabel("Countries")
plt.title("Cocoa Bean Exports by Country, 2018")
plt.savefig("Lienar_Plot_Cocoa_Beans_Export.jpg")
# plt.yscale("log")
# plt.show()
```
