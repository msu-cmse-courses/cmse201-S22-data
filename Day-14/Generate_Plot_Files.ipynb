```python
### ANSWER
# # This code was used to generate the embedded plot
import math
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

def population(time,pinitial,bigC,littlek):
    a = (bigC - pinitial)/pinitial
    pop = []
    for t in time:
        pop.append(bigC/(1 + a*math.exp(-1.0*littlek*t)))
    return pop

sns.set_style("ticks")
sns.set_context("notebook")
time = []
for i in range(300):
    time.append(i)

Cs = [8,9,10,11,12]
for C in Cs:
    p = population(time, 1, C, 0.025)
    plt.plot(time,p,label="C = "+str(C)+" billion")
plt.legend()
plt.grid()
plt.xlabel("Time [years]")
plt.ylabel("Population [billions]")
plt.title("Population models (P$_{init}$ = 1 billion, k = 0.25)")
plt.tight_layout()
sns.despine()
plt.savefig("population_models_bigC.png")
```

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

def population(time,pinitial,bigC,littlek):
    a = (bigC - pinitial)/pinitial
    pop = []
    for t in time:
        pop.append(bigC/(1 + a*math.exp(-1.0*littlek*t)))
    return pop

sns.set_style("ticks")
sns.set_context("notebook")
time = []
for i in range(300):
    time.append(i)

ks = [0.015,0.025,0.035,0.045,0.055]
for k in ks:
    p = population(time, 1, 12, k)
    plt.plot(time,p,label="k = "+str(k))
plt.legend()
plt.grid()
plt.xlabel("Time [years]")
plt.ylabel("Population [billions]")
plt.title("Population models (P$_{init}$ = 1 billion, C = 12)")
plt.tight_layout()
sns.despine()
plt.savefig("population_models_littlek.png")
```
