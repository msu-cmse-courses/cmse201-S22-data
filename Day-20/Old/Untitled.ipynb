```python
import numpy as np
# df = df.drop
values = [0]
df_nonzero = df[df.Value.isin(values) == False]


# df_nonzero.describe()
df_nz_choc = df_nonzero[df_nonzero.Item == "Chocolate products nes"]
df_nz_bean = df_nonzero[df_nonzero.Item == "Cocoa, beans"]


df_nz_choc_top = df_nz_choc.sort_values('Value',ascending=False).head(int(20))
df_nz_bean_top = df_nz_bean.sort_values('Value',ascending=False).head(int(20))


# keep_countries = df_nz_choc_top.Area.unique()
# df_nz_bean_top_keep = df_nz_bean_top[df_nz_bean_top.Area.isin(keep_countries) == True]
# print(df_nz_bean_top_keep.count())
# print(df_nz_bean_top_keep["Area"])

bean_exports, choc_export, countries = [],[],[]
country_list = np.unique(np.array(df_nz_choc_top["Area"].to_list()+df_nz_bean_top["Area"].to_list()))
# print(country_list)

total_bean_exports = df_nz_bean.Value.sum()
total_choc_exports = df_nz_choc.Value.sum()

for jj in range(len(country_list)):
# for jj in range(2):
    df_country = df[df["Area"] == country_list[jj]]
    df_temp1 = df_country[df_country["Item"] == "Cocoa, beans"]
    df_temp2 = df_country[df_country["Item"] == "Chocolate products nes"]
#     print(df_temp2.Value.to_list()[0])
    bean_exports.append(df_temp1.Value.to_list()[0]/total_bean_exports) 
    choc_export.append(df_temp2.Value.to_list()[0]/total_choc_exports) 
    #     bean_exports.append(df_country[df_country["Item"] == "Cocoa, beans"].Value.to_list()[0]) 
#     choc_export.append(df_country[df_country["Item"] == "Chocolate products nes"].Value.to_list()[0]) 
print(bean_exports)
# df_nz_choc_keep = df_nz_choc[df_nz_choc.Area.isin(keep_countries) == True]
# # df_nz_bean_keep = df_nz_choc[df_nz_bean.Value.isin(keep_countries ) == False]
# print(df_nz_choc_keep.count())
# print(df_nz_bean.count())


# for country in df["Area"].unique():
#     df_country = df[df["Area"] == country]
#     if 
```
